## Mining Champion.gg:

```python
#url = "http://api.champion.gg/v2/champions/1/MIDDLE/matchups?elo=SILVER&api_key=bfdd5436cda0189a4237f2bb6df04e74"
#page = rq.get(url)
#parsed_page = BeautifulSoup(page.text,"html.parser")

url = "pages/champion_statistic.html"
parsed_page = BeautifulSoup(open(url),"html.parser")
#parsed_page
```

```python
champions = []
nameClear = lambda x: re.sub(r' ', '', re.sub(r'\'','', x))
for a in parsed_page.find_all("tr", class_="ng-scope"):
    
    champion = nameClear(a.find("span", class_="stat-champ-title ng-binding").text)
    role = a.find("td", class_="stats-role-title ng-binding").text
    winRate = a.find_all("span", class_="ng-binding")[1].text
    playPercent = a.find_all("td", class_="ng-binding")[2].text
    banRate = a.find_all("td", class_="ng-binding")[3].text
    experience = a.find_all("td", class_="ng-binding")[4].text
    kills = a.find_all("td", class_="ng-binding")[5].text
    death = a.find_all("td", class_="ng-binding")[6].text
    assist = a.find_all("td", class_="ng-binding")[7].text
    largestKillingSpree = a.find_all("td", class_="ng-binding")[8].text
    totalDamageDealtToChampions = a.find_all("td", class_="ng-binding")[9].text
    totalDamageTaken = a.find_all("td", class_="ng-binding")[10].text
    totalHeal = a.find_all("td", class_="ng-binding")[11].text
    minionsKilled = a.find_all("td", class_="ng-binding")[12].text
    neutralMinionsKilledEnemyJungle = a.find_all("td", class_="ng-binding")[13].text
    neutralMinionsKilledTeamJungle = a.find_all("td", class_="ng-binding")[14].text
    goldEarned = a.find_all("td", class_="ng-binding")[15].text
    overallPosition = a.find_all("td", class_="ng-binding")[16].text
    
    champions.append(dict(zip(["champion","role","winRate","playPercent","banRate","experience","kills","death",\
                               "assist","largestKillingSpree","totalDamageDealtToChampions","totalDamageTaken",\
                               "totalHeal","minionsKilled","neutralMinionsKilledEnemyJungle","neutralMinionsKilledTeamJungle",\
                               "goldEarned","overallPosition"],\
                             [champion, role, winRate, playPercent, banRate, experience, kills, death, assist, \
                             largestKillingSpree, totalDamageDealtToChampions, totalDamageTaken, totalHeal, minionsKilled,\
                             neutralMinionsKilledEnemyJungle, neutralMinionsKilledTeamJungle, goldEarned, overallPosition])))
champions
```

```python
dataChampions = pd.DataFrame(champions, columns=["champion","role","winRate","playPercent","banRate","experience","kills",\
                                                 "death","assist","largestKillingSpree","totalDamageDealtToChampions",\
                                                 "totalDamageTaken","totalHeal","minionsKilled","neutralMinionsKilledEnemyJungle",\
             
                                                 "neutralMinionsKilledTeamJungle","goldEarned","overallPosition"])


dataChampions['role'] = dataChampions.role.astype('category')
dataChampions["winRate"] = dataChampions.winRate.apply(lambda x: float(x.strip('%'))/100)
dataChampions["playPercent"] = dataChampions.playPercent.apply(lambda x: float(x.strip('%'))/100)
dataChampions["banRate"] = dataChampions.banRate.apply(lambda x: float(x.strip('%'))/100)
dataChampions["experience"] = dataChampions.experience.astype('float')
dataChampions["kills"] = dataChampions.kills.astype('float')
dataChampions["death"] = dataChampions.death.astype('float')
dataChampions["assist"] = dataChampions.banRate.astype('float')
dataChampions["totalDamageDealtToChampions"] = dataChampions.totalDamageDealtToChampions.astype('int')
dataChampions["totalDamageTaken"] = dataChampions.totalDamageTaken.astype('int')
dataChampions["largestKillingSpree"] = dataChampions.largestKillingSpree.astype('int')
dataChampions["totalHeal"] = dataChampions.totalHeal.astype('int')
dataChampions["minionsKilled"] = dataChampions.minionsKilled.astype('float')
dataChampions["neutralMinionsKilledEnemyJungle"] = dataChampions.neutralMinionsKilledEnemyJungle.astype('float')
dataChampions["neutralMinionsKilledTeamJungle"] = dataChampions.neutralMinionsKilledTeamJungle.astype('float')
dataChampions["goldEarned"] = dataChampions.goldEarned.astype('int')
dataChampions["overallPosition"] = dataChampions.overallPosition.astype('int')

dataChampions.to_csv("csv/data_champions.csv")
```

## Mining championship matchs:

```python
url = "pages/championship_matchs.html"
parsed_page = BeautifulSoup(open(url),"html.parser")
```

```python
matchs = []
table = parsed_page.find_all("table", class_="wikitable")[13]
table2 = parsed_page.find_all("table", class_="wikitable")[14]
table3 = parsed_page.find_all("table", class_="wikitable")[15]

#Quartas de Final
i=0
for a in table.find_all("tr"):
    if(i>0):
        blue_team = a.find_all("a")[0].get('title')
        red_team = a.find_all("a")[2].get('title')
        match1 = a.find_all("a")[4].get('href')
        match2 = a.find_all("a")[6].get('href')
        match3 = a.find_all("a")[8].get('href')
        match4 = a.find_all("a")[10].get('href')
        match5 = a.find_all("a")[12].get('href')
        matchs.append(dict(zip(["blue_team","red_team","match1","match2","match3","match4","match5"],\
                              [blue_team, red_team, match1, match2, match3, match4, match5])))
    i=i+1
i=0

#Semi-Final
for a in table2.find_all("tr"):
    if(i>0):
        blue_team = a.find_all("a")[0].get('title')
        red_team = a.find_all("a")[2].get('title')
        match1 = a.find_all("a")[4].get('href')
        match2 = a.find_all("a")[6].get('href')
        match3 = a.find_all("a")[8].get('href')
        match4 = a.find_all("a")[10].get('href')
        match5 = a.find_all("a")[12].get('href')
        matchs.append(dict(zip(["blue_team","red_team","match1","match2","match3","match4","match5"],\
                              [blue_team, red_team, match1, match2, match3, match4, match5])))
    i=i+1
    
i=0

#Final
for a in table3.find_all("tr"):
    if(i>0):
        blue_team = a.find_all("a")[0].get('title')
        red_team = a.find_all("a")[2].get('title')
        match1 = a.find_all("a")[4].get('href')
        match2 = a.find_all("a")[6].get('href')
        match3 = a.find_all("a")[8].get('href')
        match4 = a.find_all("a")[10].get('href')
        match5 = a.find_all("a")[12].get('href')
        matchs.append(dict(zip(["blue_team","red_team","match1","match2","match3","match4","match5"],\
                              [blue_team, red_team, match1, match2, match3, match4, match5])))
    i=i+1

dataMatchs = pd.DataFrame(matchs, columns=["blue_team", "red_team", "match1", "match2", "match3", "match4", "match5"])
dataMatchs.to_csv("csv/links_matchs.csv")
```

```python
##MATCHS

stringToFloat = lambda x: float(re.sub(r'k', '', x))

def scrapingData(parsed_page):
    champion1 = parsed_page.find_all("div", class_="champion-nameplate-icon")[0].find_all("div")[1].get("data-rg-id")
    champion2 = parsed_page.find_all("div", class_="champion-nameplate-icon")[1].find_all("div")[1].get("data-rg-id")
    champion3 = parsed_page.find_all("div", class_="champion-nameplate-icon")[2].find_all("div")[1].get("data-rg-id")
    champion4 = parsed_page.find_all("div", class_="champion-nameplate-icon")[3].find_all("div")[1].get("data-rg-id")
    champion5 = parsed_page.find_all("div", class_="champion-nameplate-icon")[4].find_all("div")[1].get("data-rg-id")
    champion6 = parsed_page.find_all("div", class_="champion-nameplate-icon")[10].find_all("div")[1].get("data-rg-id")
    champion7 = parsed_page.find_all("div", class_="champion-nameplate-icon")[11].find_all("div")[1].get("data-rg-id")
    champion8 = parsed_page.find_all("div", class_="champion-nameplate-icon")[12].find_all("div")[1].get("data-rg-id")
    champion9 = parsed_page.find_all("div", class_="champion-nameplate-icon")[13].find_all("div")[1].get("data-rg-id")
    champion10 = parsed_page.find_all("div", class_="champion-nameplate-icon")[14].find_all("div")[1].get("data-rg-id")
    status1 = parsed_page.find_all("div", class_="game-conclusion")[0].text
    status2 = parsed_page.find_all("div", class_="game-conclusion")[1].text
    team1 = parsed_page.find_all(class_="champion-nameplate-name")[0].text
    team2 = parsed_page.find_all(class_="champion-nameplate-name")[5].text
    gold1 = stringToFloat(parsed_page.find_all("div", class_="gold")[0].text)
    gold2 = stringToFloat(parsed_page.find_all("div", class_="gold")[7].text)
    date = parsed_page.find("span", class_="map-header-date").text;
    time1.append(dict(zip(["champion1","champion2","champion3","champion4","champion5","status1", "team", "gold", "date"],\
                         [champion1, champion2, champion3, champion4, champion5, status1, team1, gold1, date])))
    time1.append(dict(zip(["champion1","champion2","champion3","champion4","champion5", "status1", "team", "gold", "date"],\
                         [champion6, champion7, champion8, champion9, champion10, status2, team2, gold2, date])))
    return time1
```

```python
time1 = []
times = []
def getData():
    for i in range(29):
        url = ("pages/match"+str(i+1)+".html")
        parsed_page = BeautifulSoup(open(url, encoding="utf8"), "html.parser")
        times = scrapingData(parsed_page)
    return times
        

times = getData()
dataMatchs2 = pd.DataFrame(times, columns = ["champion1", "champion2", "champion3", "champion4", \
                                             "champion5", "status1", "team", "date"])


dataMatchs2['date'] = pd.to_datetime(dataMatchs2['date'])
dataMatchs2['team'] = dataMatchs2.team.apply(lambda x: x.split(" ", 2)[1])
dataMatchs2['status1'] = dataMatchs2.status1.astype('category')
dataMatchs2.to_csv("csv/matchs.csv")
dataMatchs2
```
