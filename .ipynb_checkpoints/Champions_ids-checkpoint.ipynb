# Predição do resultado das partidas profissionais de league of legends

O objeto do projeto é minerar dados das partidas profissionais de league of legends e estatísticas dos personagens e utilizar estes dados para predizer o resultado destas partidas

Para isto utilizaremos dados de sites como http://www.probuilds.net/ para coletar as partidas dos jogadores profissionais e dados de sites como o http://champion.gg/ para coletar estatísticas dos personagens do jogo.

Inicialmente estudamos as APIs e sites que podem ser utilizados para obter os dados que precisamos.
A API do champion.gg se mostrou bastante útil para o nosso objetivo, enquanto houveram alguns problemas na mineração das informações diretamente do site.
Porém, a API do champion.gg trabalha com o ID dos personagens, desta forma foi necessário coletar os IDs na API da Riot Games(https://developer.riotgames.com/) onde foi possível obter a lista de todos os campeões e seus respectivos IDs (https://developer.riotgames.com/api-methods/#lol-static-data-v3/GET_getChampionList)

### assim criamos um dicionário com os IDs e nomes dos campeões

```python
import requests as rq
import json
```

```python
page = rq.get('https://br1.api.riotgames.com/lol/static-data/v3/champions?locale=en_US&dataById=false&api_key=RGAPI-1e2da012-4e24-46b9-84cc-aef59de97046')
data = json.loads(page.text)
data
```

```python
champions = {}
for element in data['data'].keys():
    champion_id = data['data'][element]['id']
    champions.update({champion_id : element})

champions
```
